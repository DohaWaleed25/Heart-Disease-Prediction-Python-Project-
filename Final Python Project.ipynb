```python

age_order = ['18-24', '25-29', '30-34', '35-39',
             '40-44', '45-49', '50-54', '55-59',
             '60-64', '65-69', '70-74', '75-79', '80 or older']

df['AgeCategory'] = pd.Categorical(df['AgeCategory'], categories=age_order, ordered=True)

grouped = df.groupby(['AgeCategory', 'Sex'])['HeartDisease'].value_counts(normalize=True).rename('Proportion').reset_index()
heart_patients = grouped[grouped['HeartDisease'] == 'Yes']

plt.figure(figsize=(12,6))
sns.barplot(data=heart_patients, x='AgeCategory', y='Proportion', hue='Sex')

plt.title("Heart Disease Percentage by Age and Sex")
plt.xlabel("Age Category")
plt.ylabel("Percentage with Heart Disease")
plt.xticks(rotation=45)

plt.legend(title="Sex")

plt.show()
```

We also see that the chance of infection increases with age, especially among males.

```python
df_viz = df.copy()
cols_to_convert = ["HeartDisease", "Stroke", "DiffWalking", "KidneyDisease","Diabetic"]
for col in cols_to_convert:
    df_viz[col] = df_viz[col].map({0: "No", 1: "Yes"})

cols = ["Stroke", "DiffWalking", "KidneyDisease","Diabetic"]

fig, axes = plt.subplots(1, len(cols), figsize=(18, 5))
for i, col in enumerate(cols):
    ct = pd.crosstab(df_viz[col], df_viz["HeartDisease"], normalize='index')
    ct = ct[["Yes", "No"]]
    ct.plot(
        kind='bar',
        stacked=True,
        color=["#d62728" , "#1f77b4"] ,
        ax=axes[i],

    )
    axes[i].set_title(f"Heart Disease vs {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Proportion")
    axes[i].legend(title="Heart Disease", labels=["Yes", "No"])
    axes[i].set_ylim(0, 1)



plt.tight_layout()
plt.show()
```

**Insights:**

- When there is a stroke (Yes): A relatively large percentage (over 0.35 or 35%) of these people have heart disease (red section).



- When there is difficulty walking (Yes): A much higher percentage (more than 0.2 or 20%) of these people have heart disease (red section).



- When kidney disease is present (yes): A much higher percentage (more than 0.3 or 30%) of these people have heart disease (red section).



- When diabetes is present (yes): A much higher percentage (more than 0.2 or 20%) of these people have heart disease (red section).



- To conclude the general conclusion from all the graphs: All four graphs show a clear, positive relationship between the presence of heart disease and each of the following: stroke, difficulty walking, kidney disease, and diabetes. In other words, people with any of these conditions (stroke, difficulty walking, kidney disease, diabetes) have a significantly higher incidence of heart disease than people without these conditions. This suggests that these variables may be important risk factors for heart disease.

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

df = df.dropna()

print(X.shape)
print(y.shape)
# drop feature that unimportant and ineffective columns

X = df.drop(['HeartDisease', "Race", "GenHealth", "MentalHealth"], axis=1)   # columns have low relation in heatmp
y=df['HeartDisease']
```

```python
# split dataset train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

smote = SMOTE(random_state=42)

X_train_resampled ,  y_train_resampled = smote.fit_resample(X_train, y_train)
```

**Distribution of Dataset after Resampling**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# ندمج X و y بعد resampling في داتا فريم واحد مؤقت
df_resampled = pd.concat([pd.DataFrame(X_train_resampled, columns=X_train.columns),
                          pd.Series(y_train_resampled, name='HeartDisease')], axis=1)

sns.countplot(data=df_resampled, x='HeartDisease')
plt.title("Distribution of Heart Disease after SMOTE")
plt.xlabel("Heart Disease")
plt.ylabel("Count")
plt.show()
```

**What is XGBoost?**
🔸 XGBoost stands for eXtreme Gradient Boosting
🔸 It is an advanced Boosting algorithm.

**✳️ Its features:**
Highly accurate classification.

Faster than other Boosting algorithms.

It handles imbalanced data well.

It has the ability to automatically handle missing values.

It supports parallel processing, making it very fast.

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

#model
xgb_model = XGBClassifier(random_state=42)
#train
xgb_model.fit(X_train_resampled, y_train_resampled)

# Prediction
y_pred = xgb_model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

**1.XGBoost**
Precision = 0.24 → The percentage of people who were called sick but actually were sick is small.

Recall = 0.26 → This means the model was only able to detect 26% of those who were actually infected.

**We wanted to raise the Recall and Precision for the patient category** so we used another models

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train_resampled, y_train_resampled)
y_pred_lr = lr_model.predict(X_test)

print("Logistic Regression:")
print(classification_report(y_test, y_pred_lr))

```

**2. Logistic Regression:**
Precision (class 1): 0.19 → Not high.

Recall (class 1): 0.74→ Excellent at detecting infected people.

Accuracy: 71% → Acceptable in cases of imbalanced data.

The model identifies a large number of truly infected people, but it predicts some people have the disease when they are not (this is normal due to the trade-off between Precision and Recall).

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_resampled, y_train_resampled)
y_pred_rf = rf_model.predict(X_test)

print("Random Forest:")
print(classification_report(y_test, y_pred_rf))
```

**3. Random Forest:**
Precision (class 1): 0.21 → Slightly better.

Recall (class 1): 0.19 → Poor at detecting infected people.

Accuracy: 87% → High, but not very important because the data is unbalanced.

The model predicts infected people with low accuracy and does not detect them well. It focuses more on predicting healthy people.
